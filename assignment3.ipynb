```python
df_for = pd.read_csv('yourgoodpants_formants.csv')
df_for = df_for.drop(df_for[df_for['Time_s'] == '--undefined--'].index)
df_for = df_for.drop(df_for[df_for['F1_Hz'] == '--undefined--'].index)
df_for = df_for.drop(df_for[df_for['F2_Hz'] == '--undefined--'].index)
df_for = df_for.drop(df_for[df_for['F3_Hz'] == '--undefined--'].index)
df_for = df_for.drop(df_for[df_for['F4_Hz'] == '--undefined--'].index)

plt.figure(figsize=(18, 5))
plt.plot(df_for.Time_s.astype(float), df_for.F1_Hz.astype(float), 'ro', label='F1_Hz')
plt.plot(df_for.Time_s.astype(float), df_for.F2_Hz.astype(float), 'bo', label='F2_Hz' )
plt.plot(df_for.Time_s.astype(float), df_for.F3_Hz.astype(float), 'go', label='F3_Hz' )
plt.plot(df_for.Time_s.astype(float), df_for.F4_Hz.astype(float), 'yo', label='F4_Hz' )
plt.xlabel('time(sec)')
plt.ylabel('frequency(Hz)')
plt.legend(loc='best', ncol=4, fontsize=10, frameon=True, shadow=True)


plt.show()
```
